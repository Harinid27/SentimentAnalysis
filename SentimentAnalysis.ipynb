```python
import whisper
import torch

from transformers import pipeline

from gtts import gTTS

from IPython.display import Audio

from google.colab import files
```

```python
whisper_model = whisper.load_model("base")

print("Whisper Model Loaded Successfully")
```

```python
sentiment_model = pipeline(
    "sentiment-analysis",
    model="distilbert/distilbert-base-uncased-finetuned-sst-2-english"
)

print("Sentiment Model Loaded Successfully")
```

```python
uploaded = files.upload()
```

```python
audio_file = list(uploaded.keys())[0]

print("Uploaded Audio File:")
print(audio_file)
```

```python
transcription = whisper_model.transcribe(audio_file)
```

```python
text = transcription["text"]

print("\nTranscribed Text:\n")

print(text)
```

```python
result = sentiment_model(text)

print("\nSentiment Prediction:\n")

print(result)
```

```python
sentiment = result[0]["label"]

print("\nFinal Sentiment:")
print(sentiment)
```

```python
tts = gTTS(
    text=f"The predicted sentiment is {sentiment}",
    lang='en',
    slow=False
)
```

```python
tts.save("sentiment_output.mp3")

print("Audio Output Generated Successfully")
```
